```python
import statsmodels.formula.api as smf
lm = lm = smf.ols('GIFT_AMOUNT_2018_NEW ~ GIFT_AMOUNT_2017_NEW+GIFT_AMOUNT_2016_NEW+GIFT_AMOUNT_2015_NEW+C(elem1)+C(elem2)+C(elem3)', data = GIFT_AMOUNT).fit()
lm.summary()
```
